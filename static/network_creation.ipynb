```
import pickle
import networkx as nx
```

```
with open ('/content/drive/My Drive/Colab Notebooks/SG/final project/mystuff/all_valid_responses.p', 'rb') as fp:
    valid_responses = pickle.load(fp)
```

```
gamesNet = nx.Graph()
```

```
for res in valid_responses:
  appid = list(res.keys())
  data = res[appid[0]]['data']
  gamesNet.add_node(appid[0], name = data["name"], short_description = data["short_description"], publishers=data["publishers"], categories=[cat["id"] for cat in data["categories"]], genres=[cat["id"] for cat in data["genres"]])
```

```
len(gamesNet)
```

```
genres=nx.get_node_attributes(gamesNet,'genres')
cat=nx.get_node_attributes(gamesNet,'categories')
```

```
i=0
for from_node in gamesNet.nodes():
  print(i)
  i+=1
  for to_node in gamesNet.nodes():
    if from_node != to_node:
      from_set_genres = set(genres[from_node]) 
      to_set_genres = set(genres[to_node]) 
      from_set_cat = set(cat[from_node]) 
      to_set_cat = set(cat[to_node]) 
      if len(from_set_genres.intersection(to_set_genres)) > 2 and len(from_set_cat.intersection(to_set_cat)) > 3:
        gamesNet.add_edge(from_node, to_node, common_cat=list(from_set_cat.intersection(to_set_cat)), common_genres=list(from_set_genres.intersection(to_set_genres)))
```

```
gamesNet.size()
```

```
gamesNet.size()/(len(gamesNet)*(len(gamesNet)-1)/2)
```

```
Gcc = sorted(nx.connected_components(gamesNet), key=len, reverse=True)
gamesNetConnected = gamesNet.subgraph(Gcc[0])
```

```
gamesNetConnected.size()/(len(gamesNetConnected)*(len(gamesNetConnected)-1)/2)
```

```
nx.write_gml(gamesNet, "/content/drive/My Drive/Colab Notebooks/SG/final project/gamesNet.gml")
nx.write_gml(gamesNetConnected, "/content/drive/My Drive/Colab Notebooks/SG/final project/gamesNetConnected.gml")
```
